## 2) Load Dataset + Transformer Encoder (Embeddings)

```python
from datasets import load_dataset
from sentence_transformers import SentenceTransformer

# Load a small subset of AG News for speed
dataset = load_dataset("ag_news", split="train[:200]")

# Load sentence transformer model (from Hugging Face Hub)
model = SentenceTransformer("sentence-transformers/all-MiniLM-L6-v2")

# Extract texts
texts = dataset["text"]

# Generate embeddings
embeddings = model.encode(texts, show_progress_bar=True)

print("Loaded samples:", len(texts))
print("Embedding shape:", embeddings.shape)
```

## 3) Build Vector Database (FAISS Index)

```python
import faiss
import numpy as np

# Dimension of embedding vectors
dim = embeddings.shape[1]

# Create FAISS index (L2 distance)
index = faiss.IndexFlatL2(dim)

# Add embeddings to index
index.add(np.array(embeddings))

print("FAISS index size:", index.ntotal)
```

## 4) Retrieval Function (Mini-RAG Engine)

```python
def search(query, k=3):
    """Return top-k retrieved texts for a query."""
    q_emb = model.encode([query])
    distances, indices = index.search(np.array(q_emb), k)
    return [texts[int(i)] for i in indices[0]]
```

## 5) Try It

```python
search("stock market and economy")
```
